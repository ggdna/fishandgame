```python
import ase
import matplotlib.pyplot as plt
import colorsys
from matplotlib import rcParams
rcParams['figure.figsize'] = (4,4)
rcParams['figure.dpi'] = 150
```

```python
def print_swatch_names(swatches):
    for s in swatches:
        print(s['name'])

def get_hsv(rgb_tuple):
    r, g, b = rgb_tuple
    return colorsys.rgb_to_hsv(r, g, b)[2]

def get_hex_list(swatch_name):
    
    for s in swatches:
        if s['name'] == swatch_name:
            colors = [c['data']['values'] for c in s['swatches']]
            colors.sort(key=get_hsv)
            colors = ['#{:02x}{:02x}{:02x}'.format(int(255*c[0]),
                                                   int(255*c[1]),
                                                   int(255*c[2])) for c in colors]
    return colors

def print_json(hex_list, palette_name, subset=False):
    """
    takes hex list and makes json entry
    """
    
    if not subset:
        subset = range(len(hex_list))
#         print("{")
#         print("\t\"name\": \"%s\"," %palette_name)
#         print("\t\"colors\": %s" %str(hex_list))

    for i in range(len(subset)):
        plt.text(0, i/len(hex_list), " ▄ ▄ ▄ ▄ ▄ Index %d ▓ ▓ ▓ ▓ ▓ " %i, 
                 color=hex_list[subset[i]])
    plt.axis('off')
    plt.show()

    print("{")
    print("\t\"name\": \"%s\"," %palette_name)
    print("\t\"colors\": %s" %str([hex_list[i] for i in subset]))
        
    print("}")
```

```python
swatches = ase.parse("fish_game_colors_2.ase")

print_swatch_names(swatches)
```
